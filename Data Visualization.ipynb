```python
import matplotlib.pyplot as plt
```

```python
#Scatter Plot
```

```python
# create a figure and axis
fig,ax=plt.subplots()
# scatter the sepal_length against the sepal_width
ax.scatter(data1['caloriesburned'],data1['weight_kg'])
# set a title and labels
ax.set_title('Fitness Dataset')
ax.set_xlabel('caloriesburned')
ax.set_ylabel('weight_kg')
```

```python
#Line Chart 
```

```python
# get columns to plot
columns = data1.columns.drop(['caloriesburned'])
# create x data
x_data = range(0,data1.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, data1[column])
# set title and legend
ax.set_title('Fitness Dataset')
ax.legend()
```

```python
#Histogram
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(data1['hoursofsleep'])
# set title and labels
ax.set_title('Fitness Dataset')
ax.set_xlabel('hours of sleep')
```

```python
#Bar Chart
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = data1['caloriesburned'].value_counts() 
# get x and y data 
caloriesburned = data.index
stepcount = data.values 
# create bar chart 
ax.bar(caloriesburned, stepcount) 
# set title and labels 
ax.set_title('Fitness Dataset') 
ax.set_xlabel('caloriesburned') 
ax.set_ylabel('stepcount')
```
